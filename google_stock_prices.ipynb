Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Carregando dataset

```python
df = pd.read_csv('GOOG.csv')
```

Primeiras linhas do dataset

```python
df.head()
```

Resumo técnico do dataset

```python
df.info()
```

Convertendo data de texto para data

```python
df['Date'] = pd.to_datetime(df['Date'])
```

Definindo coluna data como índice para organizar os dados cronologicamente

```python
df.set_index('Date', inplace=True)
```

Verificando dados novamente

```python
df.head()
```

```python
df.info()
```

Calculando Tetorno Diário e Médias Móveis Simples (SMA de 50 dias para verificar curto prazo e de 200 dias para verificar longo prazo)

```python
df['Daily_Return'] = df['Close'].pct_change()
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()
```

Verificando os 5 últimos dados do dataset

```python
df.tail()
```

```python
df.info()
```

Removendo dados nulos do SMA

```python
df.dropna(inplace=True)
```

Verificando 5 primeiras linhas do dataset

```python
df.head()
```

```python
df.info()
```

Criando a visualização (gráfico)

```python
plt.figure(figsize=(15, 7))
plt.plot(df.index, df['Close'], label='Preço de Fechamento', color='cyan', alpha=0.6)
plt.plot(df.index, df['SMA_50'], label='Média Móvel 50 Dias', color='orange')
plt.plot(df.index, df['SMA_200'], label='Média Móvel 200 Dias', color='magenta')
plt.title('Histórico de Preço de Fechamento e Médias Móveis - GOOGLE', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Preço de Fechamento (USD)', fontsize=12)
plt.legend()
plt.show()
```

Gráfico de Volume

```python
plt.figure(figsize=(15, 5))
plt.bar(df.index, df['Volume'], color='gray', label='Volume de Negociação')
plt.title('Volume Diário de Negociação - GOOGLE', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Volume', fontsize=12)
plt.legend()
plt.show()
```

Calculando volatilidade móvel de 50 dias

```python
df['Volatility_50d'] = df['Daily_Return'].rolling(window=50).std()
```

Gráfico de volatibilidade

```python
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['Volatility_50d'], color='purple', label='Volatilidade (50 dias)')
plt.title('Volatilidade Móvel de 50 Dias - GOOGLE', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Desvio Padrão dos Retornos', fontsize=12)
plt.legend()
plt.show()
```

Histograma dos retornos diários

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Daily_Return'], bins=100, kde=True, color='green')
plt.title('Distribuição dos Retornos Diários - GOOGLE', fontsize=16)
plt.xlabel('Retorno Diário', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.axvline(df['Daily_Return'].mean(), color='red', linestyle='--', linewidth=2, label=f"Média: {df['Daily_Return'].mean():.4f}")
plt.legend()
plt.show()
```

Dataframe limpo para csv

```python
import pandas as pd

# Carregue o SEU ARQUIVO ORIGINAL (não o que limpamos antes)
df_original = pd.read_csv('GOOG.csv')

# 1. Manter apenas as colunas essenciais
df_para_looker = df_original[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']].copy()

# 2. Converter a coluna de data para o tipo datetime
df_para_looker['Date'] = pd.to_datetime(df_para_looker['Date'])

# 3. TRANSFORMAR A DATA NO FORMATO MAIS SEGURO PARA O LOOKER (YYYYMMDD)
# Esta é a etapa mais importante para evitar erros de interpretação
df_para_looker['Date'] = df_para_looker['Date'].dt.strftime('%Y%m%d')

# 4. Salvar o novo arquivo CSV super limpo
df_para_looker.to_csv('GOOG_para_looker.csv', index=False)

print("Arquivo 'GOOG_para_looker.csv' criado com sucesso.")
print("Este arquivo contém apenas os dados brutos com a data no formato ideal.")
print(df_para_looker.head())
```
